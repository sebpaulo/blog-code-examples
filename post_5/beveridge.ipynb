### Code to reproduce the diagram illustrating the unemployment-vacancy space / beveridge curve shown in the blog post

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.00001, 5, 100)
y1 = 2 / x # rational functions as rough approximations of the Beveridge curve
y2 = 4 / x
y3 = x # 45 degree line
```

```python
fig, ax = plt.subplots(figsize=(7, 5))
ax.plot(x, y1, color="black")
ax.plot(x, y2, color="black", marker="o", markersize=1, linewidth=0)
ax.plot(x, y3, "--", color="black")
ax.annotate(
        "More efficiency",
        xy=(1.1, 1.8),
        xytext=(1.4, 2.5),
        arrowprops=dict(arrowstyle="->"),
        bbox=dict(facecolor='white', alpha=0.8)
    )
ax.annotate(
        "Less efficiency",
        xy=(2.2, 1.8),
        xytext=(1, 1),
        arrowprops=dict(arrowstyle="->"),
        bbox=dict(facecolor='white', alpha=0.8)
    )
ax.set_ylim(0, 5)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel("Unemployment rate")
ax.set_ylabel("Job vacancy rate")
ax.text(
        3, 1,
        "Recessionary",
        bbox=dict(facecolor='white', alpha=0.5, boxstyle='circle')
    )
ax.text(
        0.6, 3.4,
        "Expansionary",
        bbox=dict(facecolor='white', alpha=0.5, boxstyle='circle')
    )
ax.set_title("The Beveridge curve");
```
